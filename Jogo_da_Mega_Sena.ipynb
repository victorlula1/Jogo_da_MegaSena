```python
#Programa que simula a Mega Sena

# Adicionando bibliotecas que auxiliam na realização do sorteio e também do tempo entre a exibição da mensagens 
import random 
import time

# Mensagem e exibição das possibilidades dos números a serem sorteados na tela 
print('Você já conhece, mas vamos relembrar quais as possibilidades de números obtidos no "Jogo da MEGA SENA" \n')
time.sleep(5)
numeros_sena = set(i for i in range(1,61))
print(numeros_sena)
time.sleep(3)

# Criando variáveis que auxiliarão na captura de informações do usuário e de rotinas do programa
numeros_usuario = set()
numeros_sorteados = set()
auxiliar = 0
contador = 0
contador2 = 0
contador3 = 0

# Mensagem de início de interação com usuário 
print('\nMuito bem, agora por favor vamos aos números que você deseja escolher!')
time.sleep(3)

# Estrutura de repetição condicional 
while contador < 6:
    auxiliar = int(input('\nDigite um número: '))
    if auxiliar > 0 and auxiliar <= 60:
        if auxiliar not in numeros_usuario:
            numeros_usuario.add(auxiliar)        
            contador = contador+1 
        else:
            print('Número repetido, por favor digite outro número.')
            time.sleep(1)
    else:
        print('Por favor, digite um número entre 1 e 60')
        time.sleep(1) 
    
# Informa ao usuário em ordem crescente os números que ele apostou e os sorteados
print('\nVocê apostou nos seguintes números: ',sorted(numeros_usuario))
time.sleep(5)

print('\nCalma, muita calma nessa hora. Os números já já serão sorteados...')
time.sleep(5)

numeros_sorteados = sorted(set(random.sample(numeros_sena, 6)))

print('\nOs números sorteados na MEGA SENA foram:',numeros_sorteados)
time.sleep(3)

# Identifica quais números o usuário acertou e informa se ele ganhou ou não o prêmio de acordo com números acertados
contador2 = numeros_usuario.intersection(numeros_sorteados)
contador3 = len(contador2)

if contador3 == 6:
    print('\nVocê acertou a MEGA! Meus parabéns, você é o mais novo milionário!')
    time.sleep(3)
elif contador3 == 5:
    print('\nVocê acertou a QUINA! Meus parabéns, irá receber um ótimo prêmio!')
    time.sleep(3)
elif contador3 == 0:
    print('\nQue azar! Você não acertou nenhum número. Mais sorte da próxima vez!')
    time.sleep(3)
else:
    print('\nPoxa, você acertou o(s) número(s) {0} e que resulta em {1} número(s) certo(s). Tente uma próxima vez!'.format(contador2,contador3))
    time.sleep(3)        
print('\nFim do jogo!')

```
